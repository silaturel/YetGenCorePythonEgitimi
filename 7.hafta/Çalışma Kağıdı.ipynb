# Problem 5 Ürün Kayıt Uygulaması
Programımız ilk olarak kullanıcıdan, kaydı yapılacak ürünlerin sayısını isteyecek. Ardından yazılmış olan ürün sayısı kadar ürün ismi ve ürün fiyatı bilgisini kullanıcıdan alacak. Son olarak ise ürünleri ekrana yazdıracağız.

<h4>Sample Input:</h4>
<p>
Lutfen kac adet urun gireceginizi yaziniz: 3
<br/>
1. Urununuzun ismini giriniz: Mentos
<br/>
1. Urununuzun fiyatini giriniz: 1
<br/>
2. Urununuzun ismini giriniz: Meybuz
<br/>
2. Urununuzun fiyatini giriniz: 0.25
<br/>
3. Urununuzun ismini giriniz: Uzun Şeker
<br/>
3. Urununuzun fiyatini giriniz: 0.75
<br/>
<h4>Output:</h4>
1. Urun: Mentos - 1 TL
<br/>
2. Urun: Meybuz - 0.25 TL
<br/>
3. Urun: Uzun Şeker - 0.75 TL
</p>

```python
# 5.Problemin Çözümünü Buraya Yazınız
ürünSayisi = int(input("Lütfen kaç tane ürün gireceğinizi yazınız: "))
counter = 1
while ürünSayisi > 0:
    ürün = input(f"{counter}. ürünün ismini giriniz: ")
    fiyat = input(f"{counter}. ürünün fiyatını giriniz: ")
    print(f"{counter}. ürün: {ürün} - {fiyat} TL")
    ürünSayisi -=1
    counter +=1
```

# Problem 7 Sayı Tahmin Oyunu ama Bilgisayar Tahmin Ediyor :)
Sayı tahmin oyunu: İki oyuncudan birinin, 0 ile 100 arasında bir sayı tutması ile başlar. Diğer oyuncu ise tahminler yaparak tutulan sayıyı bulmaya çalışır ve yaptığı her tahminde sayıyı tutmuş olan oyuncudan YUKARI veya AŞAĞI olarak geri bildirim alır. Sayıyı bildiğinde ise rakip oyuncu kaç tahminde (kaç denemede) bildiğini söyler.

Senden yazmanı beklediğimiz program ise Python'ın tahminleri yapan tarafta olacağı bir oyun oluşturman. Yani sen 0 ile 100 arasında bir sayı tutacaksın ardından program senin tuttuğunuz sayıyı bulmaya çalışacak.
<h4>Sample Input:</h4>
Bu örnekte aklınızdan 90 tuttuğunuzu varsayalım.
<p>
Oyun başlıyor lutfen aklindan bir sayi tut! Tuttun mu, hazır mısın? : 1
<br>
Tahmin: 50
<br>
Durum: Yukari
<br>
Tahmin: 75
<br>
Durum: Yukari
<br>
Tahmin: 87
<br>
Durum: Yukari
<br>
Tahmin: 93
<br>
Durum: Asagi
<br>
Tahmin: 90
<br>
Durum: Bildin
</p>

```python
# 7.Problemin Çözümünü Buraya Yazınız
import random
sayi = int(input("Lütfen 0 ile 100 arasında bir sayı giriniz (0 ve 100 dahil değil): "))
tahmin = random.randint(0,100)
tahminAraligi = [0,100]
print("Tahmin:" , tahmin)
kalanHak = 5
while kalanHak > 0:
    kalanHak -= 1
    if tahmin < sayi:
        if kalanHak == 0:
            print("Kaybettin! Sayı:" , sayi)
        else:
            tahminAraligi[0] = tahmin
            print("Durum: Yukarı")
            newTahmin = random.randint(tahminAraligi[0] + 1,tahminAraligi[1] - 1)
            print("Tahmin:" , newTahmin)
            tahmin = newTahmin 
    elif tahmin > sayi:
        if kalanHak == 0:
            print("Kaybettin! Sayı:" , sayi)
        else:
            tahminAraligi[1] = tahmin
            print("Durum: Aşağı")
            newTahmin = random.randint(tahminAraligi[0] + 1,tahminAraligi[1] - 1)
            print("Tahmin:" , newTahmin)
            tahmin = newTahmin 
    else:
        print("Tebrikler! Bildin") 
        break
print("Son tahmin aralığı:" , tahminAraligi)
```

# Problem 8 Net Yaş Hesaplayıcı

Kullanıcının bugünün güncel tarihine kadar yaşadığı ay, gün, yıl süresini hesaplayan bir Python kodu yazınız. (Her ay 30 gün kabul edilmeyecektir. Kodu yazarken hangi ayın kaç gün olduğuna [Şubat ayının 29 gün olduğu yıllar dahil] dikkat edin.)

```python
# 8.Problemin Çözümünü Buraya Yazınız
import datetime

# Liste içinde doğum tarihi
dogumTarihi = input("Doğum tarihinizi GG.AA.YYYY cinsinden giriniz: ").split(".")
tarihListe = list(map(int, dogumTarihi))
print("Doğum tarihi:" , tarihListe)

# Liste içinde bugünün tarihi
bugun = list(str(datetime.date.today()).split("-"))
bugunListe = list(map(int, bugun))
bugunListe.reverse()
print("Bugünün tarihi:" , bugunListe)

# 29 Şubatları hesaplayacağımız fonksiyon
def artıkYil(x):
    if x % 4 == 0:
        if x % 400 == 0:
            return x
        elif x % 100 != 0:
            return x

# Yıl sayısı
yil = int(bugunListe[2]) - int(dogumTarihi[2])
if int(dogumTarihi[1]) > int(bugunListe[1]):
    yil -= 1 

# Ay sayısı
ay = 12 * yil
if int(dogumTarihi[1]) < int(bugunListe[1]):
    ay += (int(bugunListe[1]) - int(dogumTarihi[1]))
elif int(dogumTarihi[1]) > int(bugunListe[1]):
    ay += (12 - (int(dogumTarihi[1]) - int(bugunListe[1])))

# Hangi ay kaç gün?
gün31 = [1,3,5,7,8,10,12]
gün30 = [4,6,9,11]
gün28 = [2]

# Gün sayısı
gün = 365 * yil
for i in range(int(dogumTarihi[1]),13):
    if i in gün31:
        gün += 31
    elif i in gün30:
        gün += 30
    elif i in gün28:
        gün += 28 
for i in range(int(bugunListe[1]) + 1):
    if i in gün31:
        gün += 31
    elif i in gün30:
        gün += 30
    elif i in gün28:
        gün += 28 

# Şubat aylarını 28 gün üzerinden hesaplamıştık. Şimdi artık yılları hesaplayarak en doğru gün sayısına ulaştık.
for i in range(int(dogumTarihi[2]), int(bugunListe[2]) + 1):
    if artıkYil(i) == i:
        gün += 1


print(f"Siz doğduğunuzdan beri {yil} yıl, {ay} ay, {gün} gün oldu.")
```

# Problem 9 Tek-Çift

Kullanıcıdan 'q' tuşuna basana kadar rastgele sayılar alın. Tek olanları bir listeye, çift olanları bir başka listeye büyükten küçüğe sıralanacak şekilde atın. Listeleri ekrana yazdıran bir kod yazın.

```python
# 9.Problemin Çözümünü Buraya Yazınız
tek = []
cift = []
while True:
    sayi = input("Sayı giriniz: ")
    if sayi == "q":
        break
    if int(sayi) % 2 == 0:
        cift.append(int(sayi))
        cift.sort()
        cift.reverse() 
    else:
        tek.append(int(sayi)) 
        tek.sort()
        tek.reverse()
    break
print(tek)
print(cift)   
```
